```python
import requests
import pandas as pd
import numpy as np
import time
import sklearn
from bs4 import BeautifulSoup
from selenium import webdriver
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
```

```python
#trang chủ amazon
url = 'https://www.amazon.com/'
browser = webdriver.Chrome(executable_path='./chromedriver.exe') 
browser.maximize_window()
browser.get(url)
```

```python
#nhấp vào núp All cạnh thanh search chọn Books
option_search = browser.find_elements_by_css_selector('#searchDropdownBox option')
for option in option_search:
    if 'Books' in option.text:
        option.click()
        break
```

```python
#nhấp vào núp search tìm sau đó tìm đến mục Best Books of 2020 rồi nhấp vào See all results ở cuối
time.sleep(1)
browser.find_element_by_id('nav-search-submit-button').click()
time.sleep(1)
browser.find_element_by_link_text('Best Books of 2020').click()
time.sleep(1)
browser.find_element_by_link_text('See all results').click()
```

```python
#lưu link của những cuối sách
filelink = open('linkbook.txt', "w")
numlink = 0 #số lượng link thu được
```

Lấy 5 trang, mỗi trang 16 sản phẩm 

```python
for num_page in range(5):
    soup=BeautifulSoup(browser.page_source,'html.parser')
    for link in soup.find_all('a',{'class': 'a-link-normal a-text-normal'}):
        if link['href'][1] != 'd': #kí tự thứ 1 phải hoa thì mới đúng là link của book
            numlink = numlink + 1
            filelink.write(link['href']+'\n')
    browser.find_element_by_class_name('a-last').click()
    time.sleep(3)
browser.close()
numlink
```

```python
filelink.close()
```

```python
#click box -> choose what you want
def click_option(browser, box_option , option):
    box_option_types = browser.find_elements_by_css_selector(box_option)
    for option_type in box_option_types:
        if option in option_type.text:
            option_type.click()
            break
```

```python
phrases = []
status = []
boxs_option = ['#star-count-dropdown option','#star-count-dropdown option']
options = ['All positive','All critical']
reviewer_types = [1,0]
num_page = 10
len_min_content = 16
```

Note:
  * `phrase` : Tittle
  * `status` : type
  * `boxs_option`: css_selector box to click
  * `option` : type of comment to choose
  * `num_page`: how many page to check
  * `len_min_content`: min length of Tittle to get
        

```python
# get Tittle
def filter_content(browser, phrases, status, num_page, reviewer_type, len_min_content):
    for i in range(num_page):
        time.sleep(2)
        html_text = browser.page_source
        tree = BeautifulSoup(html_text, 'html.parser')
        reviews = tree.find_all('div',{'data-hook':'review'})
        for review in reviews:
            content = review.find('a', {'data-hook': 'review-title'}).text.strip()
            if(len(content) >= len_min_content):
                phrases.append(content)
                status.append(reviewer_type)
        browser.find_element_by_class_name('a-last').click()
```

```python
#Choose type of comment
def filter_data(browser, boxs_option , options, phrases, status, num_page, reviewer_types, len_min_content):
    for i in range(len(options)):
        click_option(browser, boxs_option[i] , options[i])
        filter_content(browser, phrases, status, num_page, reviewer_types[i], len_min_content)      

```

chạy từng link để lấy dữ liệu


```python
filelink = open('linkbook.txt', "r")

for i in range(numlink):
    link = filelink.readline()
    url = 'https://www.amazon.com' + link
    browser = webdriver.Chrome(executable_path='./chromedriver.exe') 
    browser.maximize_window()
    browser.get(url)
    # Chọn See all review trên browser
    review_page = browser.find_element_by_link_text('See all reviews').click()
    # chọn 'Verified purchase only' trong bảng lọc
    click_option(browser, '#reviewer-type-dropdown option' , 'Verified purchase only')
    # lần lượt chọn option 'All positive'và 'All critical' để lọc lấy content
    try:
        filter_data(browser, boxs_option , options, phrases, status, num_page, reviewer_types, len_min_content)
    except:
        pass
    browser.close()
    time.sleep(2)
```

```python
Data = pd.DataFrame(
    {'Para': phrases,
     'Stt': status
    }
)
```
